```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ploteo(n_archivo):
    with open(n_archivo, 'r') as archivo:
        n = 0
        signal = np.array([])
        for linea in archivo:
            if n > 2:
                vec = linea.split()
                signal = np.append(signal, int(vec[5]))
                n += 1
            else:
                n += 1
        t = np.arange(len(signal))
        fs = 1000
        ts = 1 / fs
        t = t * ts
        return t, signal
```

```python
biceps = ["biceps fuerza.txt", "biceps movimiento.txt", "biceps reposo.txt"]
gastrocnemio = ["gastrofuerza.txt", "gastromov.txt", "gastroreposo.txt"]
mano = ["mano contraccion sin fuerza.txt", "mano movimiento opuesto.txt", "mano reposo.txt"]
trapecio = ["trapecio movimiento fuerza.txt", "trapecio movimiento.txt", "trapecio normal.txt"]
triceps = ["triceps completo.txt", "triceps con fuerza.txt", "triceps en movimiento.txt", "triceps en reposo.txt"]

# Gráficas para bíceps
fig, axs = plt.subplots(3, 1, figsize=(20, 20))
count = 0
for i in biceps:
    t, signal = ploteo(i)
    axs[count].plot(t, signal)
    axs[count].set_title(i)
    axs[count].set_xlabel("Tiempo(s)")
    axs[count].set_ylabel("Amplitud")
    axs[count].grid()
    axs[count].set_ylim([0, 1200])  # Limitar eje Y de -1200 a 1200
    count += 1
plt.tight_layout()
plt.show()

# Gráficas para gastrocnemio
fig, axs = plt.subplots(3, 1, figsize=(20, 20))
count = 0
for i in gastrocnemio:
    t, signal = ploteo(i)
    axs[count].plot(t, signal)
    axs[count].set_title(i)
    axs[count].set_xlabel("Tiempo(s)")
    axs[count].set_ylabel("Amplitud")
    axs[count].grid()
    axs[count].set_ylim([0, 1200])  # Limitar eje Y de -1200 a 1200
    count += 1
plt.tight_layout()
plt.show()

# Gráficas para mano
fig, axs = plt.subplots(3, 1, figsize=(20, 20))
count = 0
for i in mano:
    t, signal = ploteo(i)
    axs[count].plot(t, signal)
    axs[count].set_title(i)
    axs[count].set_xlabel("Tiempo(s)")
    axs[count].set_ylabel("Amplitud")
    axs[count].grid()
    axs[count].set_ylim([0, 1200])  # Limitar eje Y de -1200 a 1200
    count += 1
plt.tight_layout()
plt.show()

# Gráficas para trapecio
fig, axs = plt.subplots(3, 1, figsize=(20, 20))
count = 0
for i in trapecio:
    t, signal = ploteo(i)
    axs[count].plot(t, signal)
    axs[count].set_title(i)
    axs[count].set_xlabel("Tiempo(s)")
    axs[count].set_ylabel("Amplitud")
    axs[count].grid()
    axs[count].set_ylim([0, 1200])  # Limitar eje Y de -1200 a 1200
    count += 1
plt.tight_layout()
plt.show()

# Gráficas para tríceps
fig, axs = plt.subplots(4, 1, figsize=(20, 20))
count = 0
for i in triceps:
    t, signal = ploteo(i)
    axs[count].plot(t, signal)
    axs[count].set_title(i)
    axs[count].set_xlabel("Tiempo(s)")
    axs[count].set_ylabel("Amplitud")
    axs[count].grid()
    axs[count].set_ylim([0, 1200])  # Limitar eje Y de -1200 a 1200
    count += 1
plt.tight_layout()
plt.show()
```
