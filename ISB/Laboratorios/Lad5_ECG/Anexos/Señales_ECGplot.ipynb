```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ploteo(n_archivo):
    with open(n_archivo, 'r') as archivo:
        n = 0
        signal = np.array([])
        for linea in archivo:
            if n > 2:
                vec = linea.split()
                signal = np.append(signal, int(vec[6]))  # Cambiar a vec[6] si el canal 2 está ahí
                n += 1
            else:
                n += 1
        t = np.arange(len(signal))
        fs = 1000
        ts = 1 / fs
        t = t * ts
        return t, signal
```

```python
D1 = ["EKG REPOSO 1.txt", "EJERCICIO D1.txt", "RESPIRACON D1.txt", "POST RESPIRACON D1.txt"]
D2 = ["EKG REPOSO 2.txt", "EJERCICIO D2.txt", "RESPIRACION D2 .txt", "POST RESPIRACION D2.txt"]
D3 = ["EKG REPOSO 3.txt", "EJERCICIO D3.txt", "RESPIRACION  D3 .txt", "POST RESPIRACION D3.txt"]
PRO = ["prosim60.txt", "prosim90.txt", "prosim120.txt", "prosim150.txt"]

# D1
fig, axs = plt.subplots(4, 1, figsize=(20, 20))
count = 0
for i in D1:
    t, signal = ploteo(i)
    axs[count].plot(t, signal)
    axs[count].set_title(i)
    axs[count].set_xlabel("Tiempo(s)")
    axs[count].set_ylabel("Amplitud")
    axs[count].grid()
    axs[count].set_ylim([0, 1200])  
    axs[count].set_xlim([0, 20])  
    count += 1
plt.tight_layout()
plt.show()

# D2
fig, axs = plt.subplots(4, 1, figsize=(20, 20))
count = 0
for i in D2:
    t, signal = ploteo(i)
    axs[count].plot(t, signal)
    axs[count].set_title(i)
    axs[count].set_xlabel("Tiempo(s)")
    axs[count].set_ylabel("Amplitud")
    axs[count].grid()
    axs[count].set_ylim([0, 1200])  
    axs[count].set_xlim([0, 20])  
    count += 1
plt.tight_layout()
plt.show()

# D3
fig, axs = plt.subplots(4, 1, figsize=(20, 20))
count = 0
for i in D3:
    t, signal = ploteo(i)
    axs[count].plot(t, signal)
    axs[count].set_title(i)
    axs[count].set_xlabel("Tiempo(s)")
    axs[count].set_ylabel("Amplitud")
    axs[count].grid()
    axs[count].set_ylim([0, 1200])  
    axs[count].set_xlim([0, 20])  
    count += 1
plt.tight_layout()
plt.show()

# PRO
fig, axs = plt.subplots(4, 1, figsize=(20, 20))
count = 0
for i in PRO:
    t, signal = ploteo(i)
    axs[count].plot(t, signal)
    axs[count].set_title(i)
    axs[count].set_xlabel("Tiempo(s)")
    axs[count].set_ylabel("Amplitud")
    axs[count].grid()
    axs[count].set_ylim([0, 1200])  
    axs[count].set_xlim([0, 20])  
    count += 1
plt.tight_layout()
plt.show()
```
